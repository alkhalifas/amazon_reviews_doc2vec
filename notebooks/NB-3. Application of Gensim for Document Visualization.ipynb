### Import ibraries:

```python
import pandas as pd
import numpy as np

from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize

import seaborn as sns
sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)

import matplotlib.pyplot as plt
import umap
import umap.plot

import nltk
nltk.download('punkt')
```

### Import Dataset:

```python
df = pd.read_csv("../data/Reviews.csv")
df.head()
```

### Ranomly Sample a Smaller Subset:

```python
df = df.sample(20000)
df = df.reset_index(drop=True)
df.head()
```

```python
df.shape
```

### Check Data:

```python
reviews = df["Text"].values
reviews[:5]
```

### Create TaggedDocuments:

```python
tagged_reviews = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(reviews)]
tagged_reviews[:5]
```

### Prepare Gensim Model:

```python
max_epochs = 30
vec_size = 2
alpha = 0.025
```

```python
model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1)
```

```python
model.build_vocab(tagged_reviews)
```

```python
for epoch in range(max_epochs):
    model.train(tagged_reviews,
                total_examples=model.corpus_count,
                epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
```

```python
model.save("../models/doc2vec_v1.model")
```

### Apply Model to New Data:

```python
from gensim.models.doc2vec import Doc2Vec

model = Doc2Vec.load("../models/doc2vec_v1.model")
#to find the vector of a document which is not in training data
new_data = word_tokenize("If you are looking for the secret ingredient in Robitussin I believe I have found it.  I got this in addition to the Root Beer Extract I ordered (which was good) and made some cherry soda.  The flavor is very medicinal.".lower())
vector1 = model.infer_vector(new_data)
vector1
```

### Find Similar Documents:

```python
similar_doc = model.docvecs.most_similar('1')
print(similar_doc)
```

### Find Vectors of a trained Document:

```python
# to find vector of doc in training data using tags or in other words, printing the vector of document at index 1 in training data
print(model.docvecs['1'])
```

### Vectorize All Documents:

```python
# Can be applied to new or trained documents. Trained document vectors are already saved in the model.

coords = []

def get_coords(text):
    tokens = word_tokenize(text)
    return model.infer_vector(tokens)

for text in reviews:
    coords.append(get_coords(text))
    
print(len(coords))
```

```python
df_all = pd.DataFrame(coords, columns = ["GS1", "GS2"])
df_all["Score"] = df["Score"]
df_all["ProductId"] = df["ProductId"]
df_all.head()
```

### Visualize Results:

```python
mapper = umap.UMAP(random_state=42).fit(coords)
mapper
```

```python
import umap.plot
umap.plot.points(mapper, width=500, height=500)
```

```python
umap.plot.points(mapper, labels=df_all["Score"].values)
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(coords)
X_embedded.shape
```

```python
X_embedded
```

```python
plt.figure(figsize=(15,10))
plt.title("t-SNE Graphical Representation of Amazon Reviews", fontsize=25)
sns.scatterplot(
    data = X_embedded, 
    x = X_embedded[:,0], 
    y = X_embedded[:,1],
    alpha=0.5,
    s = 20)
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(n_clusters = 50).fit_predict(X_embedded)
clustering
```

```python
plt.figure(figsize=(15,10))
plt.title("t-SNE Graphical Representation of Clustered Amazon Reviews", fontsize=25)
sns.color_palette("Blues", as_cmap=True)

sns.scatterplot(
            data = X_embedded, 
            x = X_embedded[:,0], 
            y = X_embedded[:,1],
            hue = clustering,
            alpha = 0.5,
            s = 20)
```
